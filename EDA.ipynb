```python
import pandas as pd
df = pd.read_csv("retail_sales_dataset.csv")
df.head()
df.info()
df.describe()
df.isnull().sum()
df['Date']=pd.to_datetime(df['Date'])
df.info()
df.groupby('Product Category')['Total Amount'].sum().sort_values(ascending=False)
import matplotlib.pyplot as plt
df.groupby('Product Category')['Total Amount'].sum().sort_values(ascending=False).plot(kind='bar')
plt.title('Total Amount by Product Category')
plt.xlabel('Product Category')
plt.ylabel('Total Sales')
plt.show()
df.groupby('Gender')['Total Amount'].sum()
df.groupby('Gender')['Total Amount'].sum().plot(kind='bar')
plt.title('Total Sales by Gender')
plt.xlabel('Gender')
plt.ylabel('Total Sales')
plt.show()
df['Age Group']=pd.cut(df['Age'],bins=[0,18,30,45,60,100], labels=['0-18','18-30','30-45','45-60','60+'])
age_sales= df.groupby('Age Group')['Total Amount'].sum()
age_sales
age_sales.plot(kind='bar')
plt.title('Total Sales by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Total Sales')
plt.show()
df['Month']=df['Date'].dt.month
monthly_sales= df.groupby('Month')['Total Amount'].sum()
monthly_sales
monthly_sales.plot(kind='line', marker='o')
plt.title('Total Sales by Month')
plt.xlabel('Month')
plt.ylabel('Total Sales')
plt.show()
```

# EDA INSIGHTS (RETAIL SALES)
1. Electronics Category generates the highest revenue
2. Female customers contribute to total sales
3. Age group 45-60 spend the most
4. Month 5 has the highest sales.
5. Sales vary significantly across categories and demographics
